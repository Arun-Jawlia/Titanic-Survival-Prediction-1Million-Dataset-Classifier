## Prediction using ANN Model

```python
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import tensorflow as tf
```

```python
with open("../label_encoder.pkl", 'rb') as file:
    label = pickle.load(file)
```

```python
with open("../onehot_encoder.pkl", 'rb') as file:
    onehot = pickle.load(file)
```

```python
with open("../scaler.pkl", 'rb') as file:
    scaler = pickle.load(file)
```

```python
model = load_model('../model.h5')
```

```python
user={'Pclass':3,'Sex': 'female', 'SibSp':1, 'Parch':1, 'Fare':11, 'Embarked': 'Queenstown'}
```

```python
user={'Pclass':1,'Sex': 'male', 'SibSp':1, 'Parch':1, 'Fare':11, 'Embarked': 'Queenstown'}
```

```python
user={'Pclass':1,'Sex': 'female', 'SibSp':1, 'Parch':1, 'Fare':11, 'Embarked': 'Chebourg'}
```

```python
user = pd.DataFrame([user])
```

```python
user
```

```python
user['Sex'] = label.transform(user['Sex'])
```

```python
user
```

```python
embarked = onehot.transform(user[['Embarked']])
```

```python
embarked = pd.DataFrame(embarked, columns=onehot.get_feature_names_out())
```

```python
embarked
```

```python
user = pd.concat([user.drop(columns=['Embarked']), embarked],axis=1)
```

```python
user
```

```python
user[['Pclass', 'SibSp', 'Parch', 'Fare']] = scaler.transform(user[['Pclass', 'SibSp', 'Parch', 'Fare']])
```

```python
user
```

```python
y = model.predict(user)
```

```python
y = y[0][0]
```

```python
y
```

```python
if y > 0.5:
    print("The Passenger will likely Survived in the Titanic Journey")
else:
    print("The Passenger won't Survived")
```
