## ANN Implementation

- How differ Deep Learning from Machine Learning is Model Training -> ANN Implementation
- DL sequential model -> input -> sequence of layers
- First sequence -> Input layer --> Hidden Layer --> Output Layer
- tensorflow -> sequential
- According to shape, 8 nodes in Input Layer 
- In hidden layer --> much number of nodes --> more complex more thinking compabilities

Input Layer
- 2 node * 3 weight + 3 bias  + 3 node * 3 weight + 3 bias + 1 node * 3 weight + 1 bias  = 25 parameter --> forward + backward propagation

Creating Sequential Model ( ANN Implementation)
- Sequential Network
- Dense -> Hidden Layer --> we can mention the no. of nodes inside the dense layer 
- Activation function - Sigmoid, TanH, ReLu
- Optimizer - AdaGrad, RMSProp, Adam, AdaDelta
- Loss function --> how to reduce
- Metrics --> Accuracy
- TensorBoard --> no. of epochs, how to reduce loss value

```python
import tensorflow
```

### Model Building Step

```python
(X_train.shape[1],)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
```

```python
# model = Sequential([Dense(64, input_shape=(X_train.shape[1],), activation='relu'), # First Hidden Layer
# Dense(32, activation='relu'),
# Dense(1, activation='sigmoid')
# ]) 
```

```python
model = Sequential([Dense(128, input_shape=(X_train.shape[1],), activation='relu'), # First Hidden Layer
Dense(64, activation='relu'),
Dense(32, activation='relu'),
Dense(1, activation='sigmoid')
]) 
```

```python
model.summary()
```

### Model Compilation : Optimizer, Loss, Metrics

```python
opt  = tensorflow.keras.optimizers.Adam(learning_rate=0.01)
```

```python
model.compile(optimizer=opt, loss='binary_crossentropy', metrics=['accuracy'])
```

### Early Stopping Value

```python
stopping_callback = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
```

```python
model.fit(X_train, y_train, validation_data=(X_valid, y_valid), epochs=10, callbacks=[stopping_callback])
```

```python
model.save("../model.h5")
```
