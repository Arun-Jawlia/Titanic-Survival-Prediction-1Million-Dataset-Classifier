## Data Cleaning

```python
data['Embarked'] =data['Embarked'].replace({'S':"Southampton", 'C':'Chebourg', 'Q':"Queenstown"})
```

```python
data.head()
```

```python
data.dropna(subset=['Embarked'] , inplace=True)
```

```python
data.isnull().sum()
```

```python
data.shape
```

```python
data['Fare'] = data['Fare'].astype('int8')
```

```python
data.head()
```

Three types of Encoder
1. Label Encoder --> Binary category --> Yes / No
2. Ordinal Encoder --> whenever category will have some order eg [Primary > Secondary > Tertiary]
3. OneHotEncoder --> Create separate columns based on requirement --> give equal priority to all 

## Data/Feature Encoding

### Gender --> LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label = LabelEncoder()
```

```python
data['Sex'] = label.fit_transform(data['Sex'])
```

```python
data.head()
```

### OneHotEncoder --> Embarked

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
onehot = OneHotEncoder(sparse_output=False)
```

```python
Embarked = onehot.fit_transform(data[['Embarked']])
```

```python
Embarked = pd.DataFrame(Embarked, columns=onehot.get_feature_names_out())
```

```python
Embarked
```

```python
data = pd.concat([data.drop(columns=['Embarked']),Embarked], axis=1)
```

```python
data.head()
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
```

```python
data.columns
```

```python
num_cols = ['Pclass', 'Sex', 'SibSp', 'Parch', 'Fare']
```

```python
data[num_cols] = scale.fit_transform(data[num_cols])
```

```python
data.head()
```
